### Connections

```python

with open("connections.json") as file:
     connections = json.load(file)
connections
```

```python
pprint(connections)
```

```python
#create tables for easy data viewing
blocked = pd.DataFrame({'username': [x for x in connections['blocked_users']], 
                        'Timestamp': pd.to_datetime([connections['blocked_users'].get(x) for x in connections['blocked_users']])})
blocked = blocked.sort_values('Timestamp', ascending = False)
blocked.head(10)
```

```python
closeFriends = pd.DataFrame({'username': [x for x in connections['close_friends']], 
                             'Timestamp': pd.to_datetime([connections['close_friends'].get(x) for x in connections['close_friends']])})
closeFriends = closeFriends.sort_values('Timestamp', ascending = False)
closeFriends.head(10)
```

```python
followers = pd.DataFrame({'username': [x for x in connections['followers']], 
                          'Timestamp': pd.to_datetime([connections['followers'].get(x) for x in connections['followers']])})
followers = followers.sort_values('Timestamp', ascending = False)
followers.head(10)
```

```python
following = pd.DataFrame({'username': [x for x in connections['following']], 
                          'Timestamp': pd.to_datetime([connections['following'].get(x) for x in connections['following']])})
following = following.sort_values('Timestamp', ascending = False)
following.head(10)
```

```python
#let's find my mutual followers (people i follow and also follow me back)
mutuals = pd.merge(following, followers, how = 'inner', on = 'username')
print('total number of mutual followers:', mutuals.shape[0])
mutuals.head(10)
```

```python
#im also interested in who im following but not following me back >:(
noFollowBack = pd.merge(following, followers, how = 'left', on = 'username').fillna(0)
noFollowBack = noFollowBack[noFollowBack['Timestamp_y'] == 0]
print('total number of non-follow Backs:', noFollowBack.shape[0])
noFollowBack['username'].values
```

```python
#now im interested in seeing who is following me but im not following back :P
notFollowingBack = pd.merge(followers, following, how = 'left', on = 'username').fillna(0)
notFollowingBack = notFollowingBack[notFollowingBack['Timestamp_y'] == 0]
print('total number of users im not following back:', notFollowingBack.shape[0])
notFollowingBack['username'].values
```

```python
pprint(commentData)
```

```python
comments = pd.DataFrame({'Comment to': [x[2] for x in commentData['media_comments']], 
                         'Comment': [x[1] for x in commentData['media_comments']], 
                         'Timestamp': pd.to_datetime([x[0] for x in commentData['media_comments']])
                        })
comments
```

```python
uniqueComments = comments.groupby('Comment to').size().sort_values()
uniqueComments = pd.DataFrame(uniqueComments).reset_index().rename(columns = {0:'number of comments'})
leastComment = uniqueComments[uniqueComments['number of comments'] == 1]['Comment to'].values
print("Some of the accounts i least comment on: ", leastComment[0:10])
print("The account i comment the most on: ", uniqueComments.iloc[231, 0])
uniqueComments
#Surprise! Surprise! I comment on my account the most 
```

```python
#let's see the correlation of my comments and likes
commentLikes = pd.merge(uniqueLikes, uniqueComments, how = 'inner', left_on ='username', right_on = 'Comment to')
commentLikes = commentLikes.drop(['Comment to'], axis = 1)
commentLikes
#interesting seems like the amount of likes isn't necessarly correlated with how much i comment on someone's post
```

```python
import seaborn as sns
sns.jointplot(
    data=commentLikes, 
    x='number of comments', 
    y='number of likes',
    );

#seems pretty consistent. I generally dont comment/like that many people's posts
```

```python
#notice most of the data is aggregated in the bottom left corner. let's zoom in to that region
plt.xlim(0, 15)
plt.ylim(0, 200)

plt.scatter(commentLikes['number of comments'], commentLikes['number of likes'])
plt.title('Number of Comments vs Likes for Each Account')
plt.xlabel('Comments')
plt.ylabel('Likes')
```

```python
#make a hexbin and contour map 
sns.jointplot(
    data=commentLikes, 
    x='number of comments', 
    y='number of likes',
    kind='kde');

sns.jointplot(
    data=commentLikes, 
    x='number of comments', 
    y='number of likes',
    kind='hex');
```

# Comment Sentiment Analysis

```python
print(''.join(open("vader_lexicon.txt").readlines()[:10]))
```

```python
sent = pd.read_csv("vader_lexicon.txt", sep='\t', usecols = [0,1], names = ['word', 'polarity'])
sent = sent.set_index('word')
sent.head()
```

```python
comments
```

```python
#let's start by cleaning our comment section
comments['Comment Filtered'] = comments['Comment'].str.lower() # lower case everything

punct = r'[^\w\s]'#filter out all punctuation
comments['Comment Filtered'] = comments['Comment Filtered'].str.replace(punct, ' ')
comments
```

```python
tidyComments = comments['Comment Filtered'].str.split(expand = True).stack().reset_index(level =1).rename(columns= {'level_1':'num', 0:'word'})
tidyComments

merged = tidyComments.merge(sent, how = 'left', left_on = 'word', right_index = True)
merged = merged.fillna(0)
merged = merged.reset_index().groupby('index').sum()
mergedComments = pd.merge(comments, merged, left_index=True, right_index=True)
mergedComments
comments['polarity'] = mergedComments['polarity']
comments
```

```python
print('Most negative comments:')
for x in comments.sort_values('polarity').head(10)['Comment']:
    print('\n  ', x)

#they are full of slight fustration but nothing too bad
```

```python
print('Most negative comments:')
for x in comments.sort_values('polarity', ascending = False).head(10)['Comment']:
    print('\n  ', x)

#most of the time it's expressing how much i love something lol. 
```
